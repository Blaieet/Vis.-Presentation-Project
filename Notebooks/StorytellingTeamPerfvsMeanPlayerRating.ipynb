Season 2014/15

```
import pandas as pd
import altair as alt
import numpy as np

# the url of the file in git repo keeps changing for some reason: might need to update it
url_pl1415 = "https://raw.githubusercontent.com/Blaieet/Vis.-Presentation-Project/main/Notebooks/DistsAndTeamPerfVsPlayerRating/data/2014-PremierLeague.csv?token=ARH4TB53KC3H6EVFBNR7IOLABXBR6"
#url_pl1415 = "https://raw.githubusercontent.com/Blaieet/Vis.-Presentation-Project/main/Notebooks/DistsAndTeamPerfVsPlayerRating/data/2014-PremierLeague.csv?token=ARH4TBYQVTG75NCS2EO3HS3AANMNM"
pl1415 = pd.read_csv(url_pl1415)


#outlier.rename(columns={"team_name": "Team Name"})
df_pointLC1415 = pl1415[pl1415["team_api_id"]==8197]

all_but_LC1415 = pl1415[pl1415["team_api_id"]!=8197]

points1415 = alt.Chart(all_but_LC1415).mark_circle(size=80,opacity=0.5#, color='grey'
                                          ).encode(alt.X('mean_player_rating',
                                          scale=alt.Scale(zero=False),
                                          axis=alt.Axis(title='Team mean player rating')), 
                                          alt.Y('points_per_game',
                                          scale=alt.Scale(zero=False), axis=alt.Axis(title='Points per game')),
                                          color=alt.Color('team_name', legend=alt.Legend(title="Team Name")), 
                                          tooltip="team_name").properties(
                                              title="Premier League, Season 2014/15"
                                              )
                                          
pointLC1415 = alt.Chart(df_pointLC1415).mark_circle(size=120,opacity=1, 
                                         color='red').encode(alt.X('mean_player_rating',
                                          scale=alt.Scale(zero=False),
                                          axis=alt.Axis(title='Team mean player rating')), 
                                          alt.Y('points_per_game',
                                          scale=alt.Scale(zero=False), axis=alt.Axis(title='Points per game')),
                                          #color=alt.Color('team_name'), 
                                          tooltip="team_name"#[alt.Tooltip("team_name", title='Team' ),
                                                   #alt.Tooltip("comment", title="Unexpected victory")
                                                   #]
                                                   ).properties(
                                              title="Premier League, Season 2014/15")
                                              
                
text1415a = alt.Chart({'values':[{'x': 78,'y': 1.2}]}).mark_text(
    text='The Leicester finished 14th,'
#, angle=346
).encode(
    x='x:Q', y='y:Q'
)

text1415b = alt.Chart({'values':[{'x': 78.1,'y': 1.1}]}).mark_text(
    text='barely escaping relegation...'
#, angle=346
).encode(
    x='x:Q', y='y:Q'
)


# for regression to work, you have to add interaction to the final plot (not before regression)
display(( points1415.transform_regression("mean_player_rating","points_per_game").mark_line().transform_fold(["Regression line"], as_=["Regression", "y"]).encode(alt.Color("Regression:N"))
+ points1415.interactive() 
+ pointLC1415.interactive()
+ text1415a+ text1415b))
```

```
import pandas as pd
import altair as alt

# the url of the file in git repo keeps changing for some reason: might need to update it
url_pl1516 = "https://raw.githubusercontent.com/Blaieet/Vis.-Presentation-Project/main/Notebooks/DistsAndTeamPerfVsPlayerRating/data/2015-PremierLeague.csv?token=ARH4TB6NBFMLVGP5X3FOX4LABXCBS"
#url_pl1516 = "https://raw.githubusercontent.com/Blaieet/Vis.-Presentation-Project/main/Notebooks/DistsAndTeamPerfVsPlayerRating/data/2015-PremierLeague.csv?token=ARH4TB7DCGQBWLEE35Y3I5DAANM2A"
pl1516 = pd.read_csv(url_pl1516)


#outlier.rename(columns={"team_name": "Team Name"})
df_pointLC1516 = pl1516[pl1516["team_api_id"]==8197]



#df_pointLC["comment"] = "All odds were againts the LC, but it ended up winning the league. Good team play, or just luck?"



points1516 = alt.Chart(pl1516).mark_circle(size=80,opacity=0.5#, color='grey'
                                          ).encode(alt.X('mean_player_rating',
                                          scale=alt.Scale(zero=False),
                                          axis=alt.Axis(title='Team mean player rating')), 
                                          alt.Y('points_per_game',
                                          scale=alt.Scale(zero=False), axis=alt.Axis(title='Points per game')),
                                          color=alt.Color('team_name', legend=alt.Legend(title="Team Name")), 
                                          tooltip="team_name").properties(
                                              title="Premier League, Season 2015/16"
                                              )
                                          
pointLC1516 = alt.Chart(df_pointLC1516).mark_circle(size=120,opacity=1, 
                                         color='green').encode(alt.X('mean_player_rating',
                                          scale=alt.Scale(zero=False),
                                          axis=alt.Axis(title='Team mean player rating')), 
                                          alt.Y('points_per_game',
                                          scale=alt.Scale(zero=False), axis=alt.Axis(title='Points per game')),
                                          #color=alt.Color('team_name'), 
                                          tooltip="team_name").properties(
                                              title="Premier League, Season 2015/16")
                                              
                
text1516 = alt.Chart({'values':[{'x': 73, 'y': 2}]}).mark_text(
    text='Here is our miracle winner, the Leicester City!'
#, angle=346
).encode(
    x='x:Q', y='y:Q'
)

# for regression to work, you have to add interaction to the final plot (not before regression)
display(( points1516.transform_regression("mean_player_rating","points_per_game").mark_line().transform_fold(["Regression line"], as_=["Regression", "y"]).encode(alt.Color("Regression:N"))
+ points1516.interactive() 
+ pointLC1516.interactive()
+ text1516))
```
