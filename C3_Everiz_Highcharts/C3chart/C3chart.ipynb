```python
import pandas as pd

premier15 = pd.read_csv("Vis.-Presentation-Project/Notebooks/DistsAndTeamPerfVsPlayerRating/data/2015-PremierLeague.csv")
premier15.head(14)
```

```python
import numpy as np
import c3pyo as c3

x = np.around(premier15["mean_player_rating"].values, 1).tolist()
y = np.around(premier15["points_per_game"].values, 1).tolist()
LC = [71.9, 2.1]
CH = [79.1, 1.3]

chart = c3.ScatterChart()
chart.plot(x, y, label='Teams of the Premier League (15/16)', color="gray"
          )
chart.plot([LC[0]], [LC[1]], label='Leicester City, the miracle winner!', color="green"
          )
chart.plot([CH[0]], [CH[1]], label='Chelsea, the disappointment', color="red"
          )
chart.xlabel('Team mean player rating')
chart.ylabel("Points per game")
#chart.title("Season 2015/16, Premier League")
chart.show()
```
